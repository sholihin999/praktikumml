#Membuat Dataset

```python
import pandas as pd

# Data training dari slide
data = {
    "temperatur_udara": [10, 25, 15, 20, 18, 20, 22, 24],
    "kecepatan_angin": [0, 0, 5, 3, 7, 10, 5, 6],
    "persepsi_marry": ["Dingin", "Panas", "Dingin", "Panas", "Dingin", "Dingin", "Panas", "Panas"]
}

df = pd.DataFrame(data)
df
```

#Hitung jarak ke data uji

```python
import math

# Titik uji
test_temp = 16
test_angin = 3

# Hitung jarak euclidean ke tiap baris data
jarak = []
for i, row in df.iterrows():
    d = math.sqrt((row["temperatur_udara"] - test_temp)**2 + (row["kecepatan_angin"] - test_angin)**2)
    jarak.append(d)

df["jarak"] = jarak
df.sort_values("jarak")
```

#Fungsi KNN sederhana

```python
from collections import Counter

def knn_predict(df, test_temp, test_angin, k):
    # Hitung jarak lagi (biar fungsi ini bisa dipakai berkali-kali)
    jarak = []
    for i, row in df.iterrows():
        d = math.sqrt((row["temperatur_udara"] - test_temp)**2 + (row["kecepatan_angin"] - test_angin)**2)
        jarak.append(d)

    df_temp = df.copy()
    df_temp["jarak"] = jarak

    # Urutkan berdasarkan jarak dan ambil k tetangga terdekat
    tetangga = df_temp.sort_values("jarak").head(k)

    # Lihat mayoritas kelas
    hitung_kelas = Counter(tetangga["persepsi_marry"])
    print(f"{k} tetangga terdekat:")
    print(tetangga[["temperatur_udara", "kecepatan_angin", "persepsi_marry", "jarak"]])
    print("==========================================================")
    print("Voting kelas:", hitung_kelas)
    print("==========================================================")
    # Ambil kelas dengan frekuensi terbanyak
    prediksi = hitung_kelas.most_common(1)[0][0]
    return prediksi
```

#Prediksi persepsi Marry

```python
#nilai K nya bisa diganti sesuai yang diinginkan saat ini menggunakan nilai K 5 sesuai praktikum saat itu
for k in [5]:
    pred = knn_predict(df, test_temp, test_angin, k)
    print(f"k = {k} -> Prediksi persepsi Marry nya adalah: {pred}\n")
```
