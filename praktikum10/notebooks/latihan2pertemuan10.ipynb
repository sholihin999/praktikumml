#Memasukan data

```python
import pandas as pd

data = {
    "NIM": ["TI001", "TI002", "TI003", "TI004", "TI005",
            "TI006", "TI007", "TI008", "TI009", "TI010"],
    "hasil_sebenarnya": [
        "Lulus", "Lulus", "Lulus", "Lulus", "Lulus",
        "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
    ],
    "hasil_prediksi": [
        "Lulus", "Lulus", "Lulus", "Tidak Lulus", "Tidak Lulus",
        "Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus", "Tidak Lulus"
    ]
}

df = pd.DataFrame(data)
df
```

#Encode Label ke Bentuk Biner

```python
# 1 = Lulus (positif)
# 0 = Tidak Lulus (negatif)

mapping = {"Lulus": 1, "Tidak Lulus": 0}

df["y_true"] = df["hasil_sebenarnya"].map(mapping)
df["y_pred"] = df["hasil_prediksi"].map(mapping)

df[["NIM", "hasil_sebenarnya", "hasil_prediksi", "y_true", "y_pred"]]
```

#Hitung Confusion Matrix dan Metrik Evaluasi

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

y_true = df["y_true"]
y_pred = df["y_pred"]

# labels=[1,0] agar urutan baris/kolom: [Lulus, Tidak Lulus]
cm = confusion_matrix(y_true, y_pred, labels=[1, 0])

accuracy  = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, pos_label=1)
recall    = recall_score(y_true, y_pred, pos_label=1)

print("Confusion Matrix (baris = actual, kolom = prediksi)")
print("Baris/kolom: [Lulus, Tidak Lulus]\n")
print(cm)

print("\nAccuracy  :", accuracy)
print("Precision :", precision)
print("Recall    :", recall)
```

#Confusion Matrix

```python
cm_df = pd.DataFrame(
    cm,
    index=["Actual Lulus", "Actual Tidak Lulus"],
    columns=["Pred Lulus", "Pred Tidak Lulus"]
)
cm_df
```

#Cetak Metrik dalam Persentase

```python
print(f"Accuracy  : {accuracy*100:.1f}%")
print(f"Precision : {precision*100:.1f}%")
print(f"Recall    : {recall*100:.1f}%")
```
